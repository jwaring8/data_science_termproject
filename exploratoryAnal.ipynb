```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
import os, random, math
import json
```

```python
file = r"C:\users\Jonathan\train.json"
data=pd.read_json(file,orient="columns")

#Add in a column counting the number of features
observationNum=data.index.tolist()
featureCount=[]
for i in range(0,49352):
    featureNum=len(data.iloc[i,6])
    featureCount.append(featureNum)
data['numberOfFeatures']=pd.Series(featureCount,index=observationNum)

# Get all our x values for every histogram
x1=data['bedrooms']
x2=data['price']/1000
x3=data['interest_level']
x4=data['numberOfFeatures']

#histogram of bedrooms
n, bins, patches = plt.hist(x1, 8, normed=1, facecolor='r',align="left")
plt.xlabel("Number of Bedrooms")
plt.ylabel('Probability')
plt.title('Histogram of Bedrooms')
plt.axis([0, 8, 0, .5])
plt.show()

#histogram of price
n, bins, patches = plt.hist(x2, 10000, normed=1, facecolor='g')
plt.xlabel("Price in Thousands")
plt.ylabel('Probability')
plt.title('Histogram of Price in Thousands')
plt.axis([0, 30, 0, .5])
plt.grid(True)
plt.show()

#histogram of interest level
x3.value_counts().plot(kind='bar')
plt.xlabel("Interest Level")
plt.ylabel('Count')
plt.title('Bar Chart of Interest Level')
plt.show()

#histogram of number of features
n, bins, patches = plt.hist(x4, 30, normed=1, facecolor='y',align="left")
plt.xlabel("Number of Features")
plt.ylabel('Probability')
plt.title('Histogram of Number of Features')
plt.axis([0, 30, 0, .5])
plt.show()
```
